```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df.columns = df.columns.str.strip().str.lower()
print(df.columns)
sns.countplot(x='area', data=df)
plt.title("Count map - Unemployment in India based on area")
plt.show()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("(", "").str.replace(")", "")
plt.figure(figsize=(5, 7))
sns.boxenplot(y=df["estimated_unemployment_rate_%"], color="purple")
plt.title("Boxen Plot - Estimated Unemployment Rate (%) in India")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```
